## Balanced Tree

Balanced tree meets following condition.

1. given $h$ = height, $n$ = size of the tree, $2^h-1 \leq n \leq 2^{h+1}-1$

while complete tree is balanced tree, balanced tree is not complete tree because there is one more limitation for complete tree: leaf nodes should be filled from left to the right.

## Binary Heap

Binary Heap is linked List(Priority Queue) based binary tree meets following two condition.
1. complete tree
2. Each node is greate than or equal to each of its children(max heap)
