## Merge Sort

Merge Sort is performed by following algorithm. 

1. divide list until all elements will splited into lists into len == 1 array.
2. concat two of them and compare. if left element is bigger than right element, change their position.
3. compare idx to idx. if right lists' element i is bigger than left lists' element i, change their position and concat.
4. iterate until all elements concated into one element.

```python
def perform_merge_sort(arr):
    if len(arr) == 1:
        return arr
    
    sub_element_1 = list()
    sub_element_2 = list()

    #   1. Divide the array into two sub-arrays

    for itr, item in enumerate(arr):
        if itr < len(arr) // 2:
            sub_element_1.append(item)
        else:
            sub_element_2.append(item)

    #   1.1. Recursively divide the sub-arrays into two sub-arrays.
    #        This is the recursive step.
    
    sub_element_1 = perform_merge_sort(sub_element_1)
    sub_element_2 = perform_merge_sort(sub_element_2)

    #   2. Sort the sub-arrays

    idx_to_count_1 = 0
    idx_to_count_2 = 0

    new_arr = list() 

    for idx in range(len(arr)):
        if idx_to_count_1 == len(sub_element_1):
            new_arr.append(sub_element_2[idx_to_count_2])
            idx_to_count_2 += 1
        elif idx_to_count_2 == len(sub_element_2):
            new_arr.append(sub_element_1[idx_to_count_1])
            idx_to_count_1 += 1
        #
        #   If the element in the first sub-array is greater than the element in the second sub-array,
        #   then the element in the second sub-array is the next element in the sorted array.
        #
        elif sub_element_1[idx_to_count_1] > sub_element_2[idx_to_count_2]:
            new_arr.append(sub_element_2[idx_to_count_2])
            idx_to_count_2 += 1
        else:
            new_arr.append(sub_element_1[idx_to_count_1])
            idx_to_count_1 += 1
    return arr
```

```python
arr = [1, 34, 2, 53, 123, 53, 12, 11, 67, 44]
print(perform_merge_sort(arr))
```
