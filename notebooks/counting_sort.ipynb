# Order of N Sorting.

## Assumption

Based on counting and numeric properties.

## Pros
- best not comparison-based

## Cons
- because of order of N sorting is not comparision-based, needs assumtions.

## Variants
- Radix sort
- Count sort

```python
def counting_sort(array: list):
    maximum = 0
    for i in range(len(array)):
        maximum = i
    count_list = list(0 for _ in range(0, maximum + 2))
    for i in array:
        if i != 0:
            count_list[i] += 1
    sorted_array = []
    for idx, num in enumerate(count_list):
        if count_list[idx] != 0:
            will_be_added = [idx] * num
            sorted_array += will_be_added
    return sorted_array 
```

```python
import random

x = [random.randint(0, 100) for i in range(100)]
print(counting_sort(x))
```
