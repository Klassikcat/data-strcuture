```python
import random

def do_quick_sort(array: list, pivot_idx: int = None):
    if len(array) <= 1:
        return array
    if pivot_idx is None:
        pivot_idx = random.choice(range(len(array))) 
    less_than = []
    greater_than = []
    pivot_val = array[pivot_idx]
    for val in array:
        if val == pivot_val:
            continue
        else:
            if val > pivot_val:
                greater_than.append(val)
            else:
                less_than.append(val)
    return do_quick_sort(less_than) + [pivot_val] + do_quick_sort(greater_than) # recursively return array(recursive version of selection sort)


```

```python
import random

numbers = [random.randint(1, 25) for _ in range(10)]
print(numbers)

do_quick_sort(numbers)
```

## Pros

if pivot is properly configured, then time complexity will be order of Nlog(N) -> In best condition.

## Cons

if pivot is configured not properly, then time complexity will be order of N^2 -> In worst condition.

## Usage

random choice machanism is usually used when select a pivot.
